# Volume Resizing
When training the nucleus predictor, we use 3D chunks taken from the original light and DNA channel volumes. The chunk dimensions must be bounded because of memory/computation time limitations. At the original volume scales, the chunks can contain only a small fraction of the cells, which would likely result in poor model training.  
Therefore, the light/DNA volumes are rescaled to achieve a pixel-actual distance ratio of ~0.3 px/um. At this scale, each chunk can contain 1-3 cells.
