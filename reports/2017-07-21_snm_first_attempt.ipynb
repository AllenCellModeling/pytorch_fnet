## Nuclear Mask Generation, sparse annotations
The idea is to build a model that can generate nuclear/DNA segmentations from the DNA channel image. The model is trainined on segmentation data from the Microscopy pipeline (that has passed QC). From the given cellular and nuclear segmentation data, we can classify each pixel as one of 3 classes:
    0. nucleus
    1. not nucleus
    2. unknown
In the "target" images below, these classes are colored black, gray, and white respectively.
